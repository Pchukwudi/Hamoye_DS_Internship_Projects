>**Question 11:** What is the total sum of Animal Fat produced in 2014 and 2017 respectively?

```python
#Perform a groupby sum aggregation on ‘Item’ and use the query function to isolate the Animal Fat Entry
fao_data.groupby('Item').sum().query('Item == "Animal fats"')
```

>>>> **Answer:** The total sum of Animal Fat produced in 2014 and 2017 are 209460.54 and 269617.53 tonnes respectively

**Question 12:** What is the mean and standard deviation across the whole dataset for the year 2015 to 3 decimal places?

```python
#use the describe funtion to get the summary statistics of the entire dataset and ge the mean, std for the year 2015
fao_data.describe()['Y2015']
```

>>>> **Answer:** The mean and std across the dataset for the year 2015 is 135.236 and 1603.404 respectively

**Question 13:** What is the total number and percentage of missing data in 2016 to 2 decimal places?

```python
#get the total number of entries using df.shape function or length of the dataframe
row, col = fao_data.shape
row
```

```python
#get the total number of missing values in the year 2016 
missing_data = fao_data.isnull().sum()['Y2016']
missing_data
```

```python
#calculate the percentage of missing data in 2016
percent = round((missing_data / row) * 100, 2)
percent
```

```python
print('The total number and percentage of missing data in 2016 is {} and {}% respectively'.format(missing_data, percent))
```

**Question 17:** Which of these elements had the highest sum in 2018?

**Question 19:** What is the total Import Quantity in Algeria in 2018?

```python
#perform a groupby operation on the dataset on 'Element' and 'Area'. Sum the total of all the elements, filter out Algeria 
#import quantity from the resulting dataset
fao_data.groupby(['Element', 'Area']).sum().query('Area == "Algeria"').query('Element == "Import Quantity"')
```

>>>> The total Import Quantity in Algeria in 2018 is **36238.29**
